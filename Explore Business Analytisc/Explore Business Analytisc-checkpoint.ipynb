```python
%matplotlib inline
plt.figure(figsize=(20,15))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
```

The graph shows the profit/loss of diffrent product in which copiers has most profit and tables gives most loss. But copies are
the lest sold so we have to increase its sales to maximise profit. Tables are not much sold and encounter a lot of loss so we 
have to either figure out the problem or we must drop this product.

```python
states = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
plt.figure(figsize=(40,40))
states[:25].plot(kind='bar', edgecolor = "#000000", color=['red','yellow'])
plt.title("Profit/Loss & Sales of top 25 States")
plt.xlabel('States')
plt.ylabel('Sales and Net profit/loss')
plt.grid(True)
states[25:].plot(kind='bar', edgecolor = "#000000", color=['red','yellow'])
plt.title("Profit/Loss & Sales of bottem 25 States")
plt.xlabel('States')
plt.ylabel('Sales and Net profit/loss')
plt.grid(True)
```

Texas has a good amount of sales but it is encountring a loss it may be due to discount so we have to figure it out. Stats like
west verginia, north dakota has very less sales and discount so we need to expend our market their.

```python
sub_cat = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
sub_cat.plot(kind='bar', color=['blue','orange'],edgecolor = "#000000")
```

In terms of both sales and profit phones, copiers, chairs are performing good. Here also we can see that table has decent amount of sales but profit is less so we have to deal with it.

```python
pd.DataFrame(df.groupby('State').sum())['Profit'].sort_values(ascending=True)
```
